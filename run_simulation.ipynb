```python
import gov_inertia as gi
```

```python
parameters = gi.initial_conditions
```

```python
history = gi.run_simulation(parameters, 1500)
```

```python
import pandas as pd
df = pd.DataFrame(history)
df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

plot_cols = ['N','G','H','mu_0','epsilon', 'savings','stolen_funds','tau','delta','zeta']
for col in plot_cols:
    plt.scatter(df.index, df[col])
    plt.title(col)
    # plt.yscale('log')
    plt.show()
```

```python

results = history
# Extract lists of values for plotting
G = [d['G'] for d in results]
N = [d['N'] for d in results]
H = [d['H'] for d in results]
savings = [d['savings'] for d in results]
epsilon = [d['epsilon'] for d in results]
# C_

plt.figure(figsize=(12, 8))

# Plot N vs G
plt.subplot(2, 2, 1)
plt.plot(G, N, marker='o')
plt.title('N vs G')
plt.xlabel('G')
plt.ylabel('N')

# Plot H vs G
plt.subplot(2, 2, 2)
plt.plot(G, H, 'orange', marker='o')
plt.title('H vs G')
plt.xlabel('G')
plt.ylabel('H')

# Plot savings vs G
plt.subplot(2, 2, 3)
plt.plot(G, savings, 'green', marker='o')
plt.title('Savings vs G')
plt.xlabel('G')
plt.ylabel('Savings')
# plt.yscale('log')

# Plot epsilon vs G
plt.subplot(2, 2, 4)
plt.plot(G, epsilon, 'red', marker='o')
plt.title('Epsilon vs G')
plt.xlabel('G')
plt.ylabel('Epsilon')
# plt.xscale('log')
# plt.yscale('log')
plt.tight_layout()
plt.show()
```
